```python
import pandas as pd
```

```python
df = pd.read_csv('fake_data_2.csv')
df
```

```python
dfCargo = pd.get_dummies(df['cargo'], prefix='cargo')
dfCargo
```

```python
df.iloc[:, 1:]
```

```python
df = dfCargo.join(df.iloc[:, 1:])
df
```

Label Encoder (sktlearn)

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df['sócio'] = le.fit_transform(df['sócio'])
df
```

# PCA

```python
corr = df[['idade', 'salario', 'bonus']].corr()
corr
```

```python
df.dropna(inplace=True)
df
```

```python
from sklearn.preprocessing import StandardScaler

# StandardScaler in all columns expect the last one
scaler = StandardScaler()
df.iloc[:, :-1] = scaler.fit_transform(df.iloc[:, :-1])
df
```

```python
from sklearn.decomposition import PCA


#pca = PCA(n_components='mle')
pca = PCA(0.9)

new_data = pca.fit_transform(df.iloc[:, :-1])
dfNew = pd.DataFrame(new_data)
dfNew
```

```python
print(pca.explained_variance_ratio_)
```

```python
from imblearn.under_sampling import NearMiss

X, y = NearMiss().fit_resample(df.iloc[:, :-1], df['sócio'])
```

```python
dfTeste = pd.DataFrame(X, columns=df.columns[:-1])
dfTeste['sócio'] = y
print(dfTeste)
```
