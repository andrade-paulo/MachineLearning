# Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(5, 4))
ax = fig.add_axes([1, 1,  1, 1])
linha1 = ax.plot([1, 2, 3, 4], [1, 2, 3, 4])
linha2 = ax.plot([1, 2, 3, 4], [2, 3, 4, 5])
```

```python
x = [3, 1, 3]
y = [3, 2, 1]

fig = plt.figure(figsize=(5, 4))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

ax.plot(x, y)
ax.set_xlim(1, 2)
ax.set_xticklabels(['mamãe', 'papai', 'vovó', 'vovô', 'titio', 'titia'])

plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.axis([0, 6, 0, 20])

plt.show()
```

```python
X, Y = range(1, 101), [x ** 2 for x in range(1, 101)]

fig1, f1_axes = plt.subplots(ncols=2, nrows=2, figsize=(6, 6))
fig.suptitle('Meus graficos', size=14)
f1_axes[0, 0].plot(X, Y, label = 'valores', color = 'blue')
f1_axes[0, 0].set_title('grafico 1')
f1_axes[0, 0].set_xlabel('label eixo x')
f1_axes[0, 0].set_ylabel('label eixo y')
f1_axes[0, 0].legend()
```

# Seaborn

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fake_doi = pd.read_csv('fake_data_2.csv')
fake_doi.head()
```

```python
print(fake_doi['idade'].sort_values())
```

```python
media = fake_doi['idade'].mean()
mediana = fake_doi['idade'].median()
moda = fake_doi['idade'].mode()

print('media:', media)
print('mediana:', mediana)
print('moda:', moda)
```

```python
plt.boxplot(fake_doi['idade'])
plt.title('Boxplot da idade')
plt.ylabel('Idade')

plt.show()
```

```python
sns.boxplot(y='idade', data=fake_doi)
# Show values of quartiles on the boxplot
q1 = fake_doi['idade'].quantile(0.25)
q2 = fake_doi['idade'].quantile(0.5)
q3 = fake_doi['idade'].quantile(0.75)

plt.text(0.01, q1*1.02, f'Q1: {q1}')
plt.text(0.01, q2*1.02, f'Q2: {q2}')
plt.text(0.01, q3*1.02, f'Q3: {q3}')
plt.show()
```

```python
# Idade por cargos
sns.boxplot(x='cargo', y='idade', data=fake_doi)
plt.show()
```

```python
sns.boxplot(x='sócio', y='salario', data=fake_doi, linewidth=2.5, hue='sócio', palette='Set2')
plt.show()
```

```python
sns.histplot(fake_doi['idade'], bins = 5, kde=True, element='bars')
```

```python
# Correlaçã
fake_doi.drop(['cargo', 'sócio'], axis='columns').corr()
```

```python
plt.scatter(fake_doi['idade'], fake_doi['bonus'])
```

```python
sns.scatterplot(x='idade', y='bonus', data=fake_doi, hue='sócio')
```
