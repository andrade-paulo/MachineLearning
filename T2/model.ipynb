```python
import pandas as pd
```

```python
data = pd.read_csv('ProcessedData/br_cities_01.csv')
data.head()
```

```python
# Extra trees with cross validation
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import cross_val_score

target_column = 'IDHM'
features = data.columns.drop(target_column)
target = data[target_column]

for n in range(20, 80, 10):
    model = ExtraTreesRegressor(n_estimators=n, max_depth=None, min_samples_split=2, min_samples_leaf=1)
    scores = cross_val_score(model, data[features], target, cv=5, scoring='r2')

    print(f"N{n}: {scores.mean()}")
```
